### Q. 각 열의 NaN 찾기, 각 열의 NaN 개수 반환, 데이터프레임의 총 NaN 개수를 반환하는 사용자 함수를 정의한 후 df.pipe 메소드로 함수 매핑한 결과를 모두 출력하세요.

```python
def total(df):
    return df.isna().sum().sum()
def count(df):
    return df.isna().sum()
def find(df):
    return df.isna
print(df.pipe(count))
print(df.pipe(find))
print(df.pipe(total))
```
