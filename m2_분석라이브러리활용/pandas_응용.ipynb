```python
import pandas as pd
file_data = pd.read_csv('file_data.csv')
file_data['가격'] = [5000,5000,4000,5000,3000]
file_data
```

```python
df1 = file_data.copy()
df2 = pd.DataFrame({
    '차림표':['비빔밥','돈까스','라면','김밥','제육덮밥'],
    '가격':[7000,8000,3000,3000,9000]
})
df2
```

```python
# 병합 - 세로방향
dfc1 = pd.concat([df1,df2])
dfc1
```
