```python
def insertcol(t):
    conn=cx_Oracle.connect('c##madang/madang@localhost:1521/xe')
    cursor = conn.cursor()
    sql = "insert into sample2 (col01,col02,col03) values (:1,:2,:3)"
    cursor.executemany(sql,t)
    cursor.close()
    conn.commit()
    conn.close()
    
conn=cx_Oracle.connect('c##madang/madang@localhost:1521/xe')
cursor = conn.cursor()
cursor.execute('drop table sample2')
cursor.execute("create table sample2(col01 varchar2(10), col02 varchar2(10), col03 varchar2(10))")

for i in range(10):
    rows=[(i,i*2,i*i)]
    insertcol(rows)


cursor.close()
conn.commit()
conn.close()
```

```python
#Update
def updatecol(t):
    conn=cx_Oracle.connect('c##madang/madang@localhost:1521/xe')
    cursor = conn.cursor()
    sql = "update sample2 set col03=:1 where col01=:2"
    cursor.execute(sql,t)
    cursor.close()
    conn.commit()
    conn.close()

updatecol(('3','1'))

#col01이 '1'이면  col03이 '3'으로 업데이트
```

```python
#delete
def deletecol(t):
    conn=cx_Oracle.connect('c##madang/madang@localhost:1521/xe')
    cursor = conn.cursor()
    sql = "delete from sample2 where col01=:1"
    cursor.execute(sql,t)
    cursor.close()
    conn.commit()
    conn.close()

deletecol(('1'))
#col01이 '1'이면 delete
```

```python
%%time
import cx_Oracle
import numpy as np
import pandas as pd

#고객별 2014년 2015년 총구매액 및 평균구매액 출력
conn=cx_Oracle.connect('c##hr/hr@localhost:1521/xe')
cursor = conn.cursor()
#'SELECT 고객번호, YEAR, SUM(구매금액),round(avg(구매금액)) FROM PURPROD GROUP BY 고객번호, YEAR ORDER BY 고객번호, YEAR'
cursor.execute('SELECT * FROM TFT')
x = cursor.fetchall()

cursor.execute("select column_name from user_tab_columns where table_name= 'TFT'")
y = cursor.fetchall()
colst=[]
for i in y:
    for j in i:
        colst.append(j)


df = pd.DataFrame(x, columns=colst)
df
```

```python
#고객별 2014 VS 2015 총구매액 기준 성장율

#고객번호 중복제거
dfn=df.drop_duplicates(['고객번호'])
dfn
```

```python
#년도별 그룹화
grp=df.groupby('YEAR')
df14=grp.get_group(2014)
df15=grp.get_group(2015)

df14
```

```python
#고객별 2014 VS 2015 총구매액 기준 성장율

#2014,2015 사이즈가 다름으로 중복제거한 고객번호 기준으로 left join
df=pd.merge(dfn,df14, on='고객번호',how='left')
df=pd.merge(df,df15, on='고객번호',how='left')
df
```

```python
df=df.loc[:,['고객번호','구매합계_y','구매합계']]
df.columns=['고객번호','2014구매합계','2015구매합계']

df
```

```python
print(df.isnull().sum())
print()
print(df.size)
```
