```python
import random

while 1:
    print("""
원하는 장르를 고르시면 총 5문제가 출제 됩니다.
1.넌센스
2.수도
3.영화 자음""")

    a=input("번호 : ")
    
    if a not in ('1','2','3') : print('■올바른 숫자를 입력해주세요.\n\n');continue
    
    with open(f'문제{a}.txt', 'r', encoding="utf-8") as f: questions = f.readlines(); random.shuffle(questions)
    
    print('=-='*5,'퀴즈시작','=-='*5)

    count = 0
    score = 0

    for s in questions:
        count += 1
        tmp = s.strip().split('\t')
        
        print(str(count) + '번 문제 : ' + tmp[0])
        
        if input('답 : ') == tmp[1]:
            print('◎ 정답!\n');score += 1
        else:
            print("X 틀렸습니다! 정답은 '" + tmp[1] + "' 입니다!\n")
        
        if count==5:break

    print(f'총 맞춘 갯수 : {score}\n\n')


    #print("""
    # 0개 아쉬움
    # 1~2개 잘했어요 
    # 3~4개 평균이상
    # 5개  전문가
    # """)
```
