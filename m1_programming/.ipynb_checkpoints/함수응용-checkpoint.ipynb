```python
def fac(n):
    base=1
    for i in range(1,n+1):
        base*=i
    return base

fac(5)
```

```python
def factorial(n):
    if n == 1:   
        return 1 
    return n * factorial(n - 1)
 
print(factorial(5))
```

```python
def cir(n):
    l=3.14*2*n
    s=l*n/2
    return l,s
print(cir(5))
```

```python
def getT(fn):
    with open(fn,'r') as f:
        text=f.read()
        fa={}
        for c in text:
            if c in fa:
                fa[c] +=1
            else:
                fa[c] =1
    return fa
ret = getT('data/mydata.txt')
ret= sorted(ret.items(),key = lambda x:x[1],reverse=True)
print(ret)
```

```python
for c,f in ret:
    if c == "\n":
        continue
    print(f'[{c}]->[{f}]회 나타남')
```

```python
for i in range(1880,2016):
    with open(f'data/names/yob{i}.txt','r') as f:
        data=f.readlines()
        sum=0
        for j in data:
            tmp=j.strip().split(',')
            tmp=int(tmp[2])
            sum+=tmp
        print(f'{i}년도 {sum}명')    
            
```

```python
for i in range(1880,2016):
    with open(f'data/names/yob{i}.txt','r') as f:
        data=f.readlines()
        man=0
        wman=0
        for j in data:
            tmp=j.strip().split(',')
            if (tmp[1]) == ('M'):
                tmp=int(tmp[2])
                man+=tmp
            elif (tmp[1]) == ('F'):
                tmp=int(tmp[2])
                wman+=tmp
            
        print(f'{i}년도 남:{man}명 여:{wman}명')    
        
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv(f'data/names/yob{i}.txt', sep = ",", encoding = "utf-8", names=['name','sex','count'])
for i in range(1880,2015):
    df1 =pd.read_csv(f'data/names/yob{i+1}.txt', sep = ",", encoding = "utf-8", names=['name','sex','count'])
    df= pd.concat([df,df1])

df = df.groupby(by=['name','sex']).sum().groupby(level=[0,1]).cumsum()
df = df.sort_values(by=df.columns[0],ascending=False)
df

```
