```python
import pandas as pd
df=pd.read_csv('C:/1차data/매출횟수등급2.csv',encoding='cp949')
df
```

```python
df.columns
ddf=df[['1기구매등급', '2기구매등급', '3기구매등급', '4기구매등급', '5기구매등급', '6기구매등급',
       '7기구매등급']]
```

```python
# display(ddf)
import numpy as np
cv=round(ddf.mean(axis=1),2)/round(ddf.std(axis=1),2)
cv
```

```python
cv.replace([np.inf, -np.inf],0,inplace=True)
cv
```

```python
ddf['변동계수']=cv
```
