```python
%%time
import pandas as pd
import warnings
warnings.simplefilter("ignore")
df = pd.read_excel('C:/1차data/16세부품목.xlsx', engine="openpyxl")
df
```

```python
pdf = pd.pivot_table(df,
                     index = '고객번호', # 행 위치에 들어갈 열
                     columns = ['분기','상품분류'], # 열 위치에 들어갈 열
                     values = ['총구매액'], # 데이터로 사용할 열
                     aggfunc = 'max') # 데이터 집계 함수
pdf
```

```python
pdf.fillna(0,inplace=True)
pdf
```

```python
pdf['총구매액']
```

```python
import numpy as np
lst=[]
for i in pdf.columns[:14]:
    lst.append(i[2])
lst
```

```python
colum=[]
for k in range(1,7):
    for j in (lst):
        colum.append(f'{k}기{j}')
colum
```

```python
col=[]
for i,j in enumerate(lst):
    for k in range(1,7):
        col.append(f'{k}기{j}')
col
```

```python
pdf.columns=colum
dfn=pdf[col]
dfn
```

```python
lst
```

```python
dfn.fillna(0,inplace=True)
```

```python
for i in range(14):
    dfa=dfn.iloc[:,0+(6*i):6+(6*i)]
    dfn[f'{lst[i]}cv']=round(dfa.mean(axis=1),2)/round(dfa.std(axis=1),2)
dfn
```

```python
res=dfn.iloc[:,-14:]
res.fillna(0,inplace=True)
```

```python
res
```

```python
dd=pd.read_csv('C:/1차data/1-7고객수.csv',encoding='utf-8')
dd
```

```python
res=pd.merge(dd,res, on='고객번호',how='left')
res
```

```python
res.to_csv('C:/1차data/1-6품목변동계수.csv',encoding='cp949',index=False)
```
