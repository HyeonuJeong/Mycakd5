```python
%%time
import pandas as pd
import warnings
warnings.simplefilter("ignore")
df1 = pd.read_excel('C:/1차data/on15.xlsx', engine="openpyxl")
df1
```

```python
pdf1 = pd.pivot_table(df1,
                     index = '고객번호', # 행 위치에 들어갈 열
                     columns = ['분기','상품분류'], # 열 위치에 들어갈 열
                     values = ['구매횟수','총구매액'], # 데이터로 사용할 열
                     aggfunc = 'max') # 데이터 집계 함수
pdf1
```
