```python
import pandas as pd
import numpy as np


budf=pd.read_excel('C:/1차data/분기별.xlsx')
badf=pd.read_excel('C:/1차data/반기별.xlsx')
codf=pd.read_excel('C:/1차data/분류코드별.xlsx')
yedf=pd.read_excel('C:/1차data/년도별.xlsx')
```

```python
pbudf = pd.pivot_table(budf,
                     index = '고객번호', # 행 위치에 들어갈 열
                     columns = '분기', # 열 위치에 들어갈 열
                     values = ['구매횟수','총구매액','구매평균'], # 데이터로 사용할 열
                     aggfunc = 'max') # 데이터 집계 함수
pbadf = pd.pivot_table(badf,
                     index = '고객번호', # 행 위치에 들어갈 열
                     columns = '반기', # 열 위치에 들어갈 열
                     values = ['구매횟수','총구매액','구매평균'], # 데이터로 사용할 열
                     aggfunc = 'max') # 데이터 집계 함수
pcodf = pd.pivot_table(codf,
                     index = '고객번호', # 행 위치에 들어갈 열
                     columns = '분류코드', # 열 위치에 들어갈 열
                     values = ['구매횟수','총구매액','구매평균'], # 데이터로 사용할 열
                     aggfunc = 'max') # 데이터 집계 함수
pyedf = pd.pivot_table(yedf,
                     index = '고객번호', # 행 위치에 들어갈 열
                     columns = 'YEAR', # 열 위치에 들어갈 열
                     values = ['구매횟수','총구매액','구매평균'], # 데이터로 사용할 열
                     aggfunc = 'max') # 데이터 집계 함수


display(pbudf)
display(pbadf)
display(pcodf)
display(pyedf)
```

```python
pbudf = pd.pivot_table(budf,
                     index = '고객번호', # 행 위치에 들어갈 열
                     columns = '분기', # 열 위치에 들어갈 열
                     values = ['구매평균'], # 데이터로 사용할 열
                     aggfunc = 'max') # 데이터 집계 함수
pbadf = pd.pivot_table(badf,
                     index = '고객번호', # 행 위치에 들어갈 열
                     columns = '반기', # 열 위치에 들어갈 열
                     values = ['구매평균'], # 데이터로 사용할 열
                     aggfunc = 'max') # 데이터 집계 함수
pcodf = pd.pivot_table(codf,
                     index = '고객번호', # 행 위치에 들어갈 열
                     columns = '분류코드', # 열 위치에 들어갈 열
                     values = ['구매평균'], # 데이터로 사용할 열
                     aggfunc = 'max') # 데이터 집계 함수
pyedf = pd.pivot_table(yedf,
                     index = '고객번호', # 행 위치에 들어갈 열
                     columns = 'YEAR', # 열 위치에 들어갈 열
                     values = ['구매평균'], # 데이터로 사용할 열
                     aggfunc = 'max') # 데이터 집계 함수


display(pbudf)
display(pbadf)
display(pcodf)
display(pyedf)
```
