```python
import pandas as pd
df = pd.read_csv('C:/1차data/RFM테이블.csv', encoding="cp949")
df
```

```python
import numpy as np

aa=df['구매빈도'].sort_values(ascending=True)
cuts = pd.qcut(aa,4,labels=np.arange(1,5))
cc=pd.concat([aa,cuts],axis=1)
cc
```
