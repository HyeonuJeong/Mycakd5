```python
# [과제] t_df의 Age 컬럼을 이용하여 새로운 컬럼 Age_cat를 사용자 함수를 이용하여 출력하세요(7개 카테고리)

def age_cat(age):
    cat=''
    if age <= 4: cat='Baby'
    elif age <= 9: cat='Child'
    elif age <=19: cat='Teenager'
    elif age <=30: cat='Young_adult'
    elif age <=60: cat='Adult'
    else: cat='Elderly'
    return cat

df['age_cat'] = df.Age.apply(lambda x: age_cat(x))
```

```python
# [과제] t_df에 대하여 결손값 처리, 불필요한 컬럼 삭제, 인코딩 등 을 수행하는 전처리 과정을 
# 하나의 사용자 함수(def transform_features(df))로 작성하여 분석용 데이터셋을 만드세요.
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

df=pd.read_csv('train.csv')

def transform_features(df):
    #결손값처리
    df['Age'].fillna(df['Age'].mean(), inplace=True)
    df['Cabin'].fillna('A', inplace=True)
    df['Embarked'].fillna('A', inplace=True)
    
    #컬럼삭제
    df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
    
    
    #인코딩
    col = ['Cabin', 'Sex','Embarked']
    df['Cabin'] = df['Cabin'].str[:1]
    for f in col:
        le = LabelEncoder() 
        le.fit(df[f])
        df[f] = le.transform(df[f])
        
        
    return df

transform_features(df)
```
